### Write a program to find if a number is a prime number ?

* It should have a command prompt asking the user to enter a number
* This number should be taken as input 
* The output should be if the number is a prime number..


```python
def isprime(num):
    if num == 0:
        return False
    if num != 2 and num % 2 == 0:
        return False
    for i in range(3, int(num / 2)):
        if num % i == 0:
            return False
    return True
```

```python
number = int(input("Enter a number : "))
result = isprime(number)
if result:
    print(f"The given number {number} is prime number.")
else:
    print(f"The given number {number} is not a prime number.")
```

##  Read a file containing numbers. The file is a text file.
* Input.txt 
* Read this file 
* Store the numbers in a data structure
* Arrange the numbers in increasing order, decreasing order.
* Find the max, min of the numbers
* Find the mean, median, mode of the numbers
* Group the numbers into a set of 10 numbers 
* For the group of 10 numbers, arrange the number in increasing or decreasing order 
* Repeat steps 5 and 6 for the set of 10 numbers

```python
with open('numbers.txt', 'r') as file:
    file_input = file.read()
    numbers = file_input.split('\n')
    for i in range(len(numbers)):
        numbers[i] = int(numbers[i])
```

```python
def sortInAscending(arr):
    for i in range(0, len(arr)):
        for j in range(i + 1, len(arr)):
            if arr[i] > arr[j]:
                arr[i], arr[j] = arr[j], arr[i]
    return arr
```

```python
def sortInDescending(arr):
    for i in range(0, len(arr)):
        for j in range(i + 1, len(arr)):
            if arr[i] < arr[j]:
                arr[i], arr[j] = arr[j], arr[i]
    return arr
```

```python
def findMax(numbers):
    max = numbers[0]
    for num in numbers:
        if max < num:
            max = num
    return max
```

```python
def findMin(numbers):
    min = numbers[0]
    for num in numbers:
        if min > num:
            min = num
    return min
```

```python
def findMean(arr):
    sum = 0
    for num in arr:
        sum += num
    return sum / len(arr)
```

```python
def findMedian(arr):
    sortInAscending(arr)
    if len(arr) % 2 == 1:
        median = arr[len(arr)//2]
    else:
        median = (arr[len(arr) // 2] + arr[len(arr) // 2 + 1]) / 2
    return median
```

```python
def findMode(arr):
    freq = dict()

    for num in arr:
        freq[num] = freq.get(num, 0)+1
        

    maxFreq = 0
    mode = 0

    for key, value in freq.items():
        if maxFreq < value:
            maxFreq = value

    for key in freq.keys():
        if maxFreq == freq[key]:
            mode = key
            break
            

    return mode
```

```python
def printResults(numbers):
    print("\n----------------------------------------------------")
    print("Numbers in ascending ", sortInAscending(numbers))
    print("Numbers in descending ", sortInDescending(numbers))    
    print("Minimun value", findMin(numbers))    
    print("Maximum value ", findMax(numbers))
    print("Mean value ", findMean(numbers))        
    print("Median value ", findMedian(numbers))        
    print("Mode value ", findMode(numbers))        
```

```python
printResults(numbers)
```

```python
def createGroups(numbers):
    groups = []
    group = []
    for i in range(len(numbers)):
        if i != 0 and i % 10 == 0:
            groups.append(group)
            group = []
        group.append(numbers[i])
    return groups

    
```

```python
groups = createGroups(numbers)
for group in groups:
    printResults(group)
```

## Ball collisions problems 
* Assume you have a two balls, with Radius “r” located at points (x1, y1) and (x2,y2). So we have 
    B1 = (x1,y1,r1), B2 = (x2,y2,r2)
### Write a program 
* That takes (x1,y1,x2,y2,r1,r2) as an input from the command line
* Calculate the relative distance between the two balls
* Find out if the balls are colliding 
* If the balls are travelling at a speed of “5 cms per second” towards earch other, print the time it will take for them to collide



```python
import math
```

```python
def calcrelativedistance(x1, x2, y1, y2):
    return math.sqrt(math.pow((x1 - x2), 2) + math.pow((y1 - y2), 2))
```

```python
def calcdistancebetweenballs(relativedistance, r1, r2):
    return relativedistance - (r1 + r2)
```

```python
def willcollide(distance):
    if distance == 0:
        print("The Balls are colliding")
    elif distance < 0:
        print("The Balls are overlapped")
    else:
        print("The Balls are not colliding")
```

```python
def calctimetocollide(distance, speed = 5):
    time = distance / speed
    if time == 0:
        print("The balls are colliding")
    elif time > 0:
        print(f"The time for balls to collide is {time} seconds")
    elif time < 0:
        print(f"The balls are already collided")
```

```python
x1 = int(input("Enter the value of x1 : "))
y1 = int(input("Enter the value of y1 : "))

x2 = int(input("Enter the value of x2 : "))
y2 = int(input("Enter the value of y2 : "))

r1 = float(input("Enter the radius for ball 1 in cm : "))
r2 = float(input("Enter the radius for ball 2 in cm : "))
```

```python
relativedistance = calcrelativedistance(x1, x2, y1, y2)
distance = calcdistancebetweenballs(relativedistance, r1, r2)
print(f"The relative distance between the balls is {relativedistance} cm")
willcollide(distance)

calctimetocollide(distance)
```

## Text or String processing
* Take any random input text file
* From the text file
* Count and print the number of words and characters
* Count the number of distinct words
* Find the frequency of distinct words 
* Reverse every line. The delimiter is a full stop. 
* Find the palindrome in the text


```python
def isPalindrome(text):
    if len(text) == 0:
        return False
    if len(text) == 1:
        return True
    for i in range(len(text)//2):
        if text[i].lower() != text[len(text)-1-i].lower():
            return False
    return True
```

```python
def reverse(text):
    sentence = ""
    words = text.split(' ')
    for i in range(len(words)):
        sentence += words[len(words)-1-i]+" "
    sentence = sentence.strip() +"."
    return sentence
```

```python
with open('file.txt', 'r', encoding='utf-8') as file:
    file_input = file.read()
    sentences = file_input.strip().split(".")
    words = []
    characters = []
    distinct = dict()
    for i in range(len(sentences)):
        sentences[i] = sentences[i].strip()
        for word in sentences[i].split(' '):
            if len(word) != 0 and word != '"':
                words.append(word.strip())
                distinct[word.lower()] = distinct.get(word.lower(), 0)+1
            for ch in word:
                characters.append(ch)
  
```

```python
print("Number of words:", len(words))
print("Number of characters:", len(characters))
print("Number of distinct words:",len(distinct.keys()))
```

```python
for sentence in sentences:
    sentence = reverse(sentence)
    print(sentence, end=" ")
```

```python
palindromes = []
for word in distinct.keys():
    if isPalindrome(word):
        palindromes.append(word)
print("The palindromes are", palindromes)
```
