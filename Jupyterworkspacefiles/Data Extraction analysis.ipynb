```python
import subprocess

def extract_text_with_pdfbox(pdf_path):
    # Path to PDFBox jar file
    pdfbox_jar_path = "D:\Downloads\pdfbox-app-3.0.2.jar"
    
    command = [
        "java", "-jar", pdfbox_jar_path, "export:text", pdf_path
    ]
    
    try:
        result = subprocess.run(command, capture_output=True, text=True, check=True)
        extracted_text = result.stdout
        return extracted_text.strip()
    except subprocess.CalledProcessError as e:
        print(f"Error executing PDFBox command: {e}")
        return None

if __name__ == "__main__":
    pdf_path = r"C:\Users\TP_USER\Downloads\POReaderFormats\POReaderFormats\Walmart\1712733040_29000070938720240410124104PM.PDF"
    extracted _text = extract_text_with_pdfbox(pdf_path)
    
    if extracted_text:
        print(extracted_text)
    else:
        print("Failed to extract text from PDF.")
```

```python
import re
from PyPDF2 import PdfReader
import os

pdfs_dir = r"C:\Users\TP_USER\Downloads\POReaderFormats\POReaderFormats"


def detect_pdf_version(pdf_path):
    with open(pdf_path, 'rb') as file:
        file_header = file.read(1024)

        version_match = re.search(rb"%PDF-(\d\.\d)", file_header)

        if version_match:
            pdf_version = version_match.group(1).decode()
            return pdf_version
        else:
            return "Unknown"


def extract_from_pdf(pdf_path):
    try:
        pdf_read = PdfReader(pdf_path)
        text = '\n'.join([page.extract_text() for page in pdf_read.pages])
        print(pdf_path)
#         print(text[:100])
        if not text:
            print("!!!!!!!\n\nNo data found\n\n!!!!!!")
        return text
    except Exception as e:
        print(f"Error processing {pdf_path}: {e}")
pdf_v_dicts = {'1.0':0, '1.1':0, '1.2':0, '1.3':0, '1.4':0, '1.5':0, '1.6':0, '1.7':0, "Unknown":0 }
all_found = 0
all_not_found = 0
for folder in os.listdir(pdfs_dir):
    data_found = 0
    not_found = 0
    directory_path = os.path.join(pdfs_dir, folder)
#     print("\n", directory_path)

    for pdf in os.listdir(directory_path):
        if pdf.lower().endswith("pdf"):
            pdf_path = os.path.join(directory_path, pdf)
            version = detect_pdf_version(pdf_path)
            print(f"\n\nPDF Version:{version}___________________________________________________________")
            text = extract_from_pdf(pdf_path)
            
            if text:
                data_found+=1
                pdf_v_dicts[version]+=1
            else:
                not_found += 1
    all_found += data_found
    all_not_found += not_found
    print(f"Total PDFs in {folder}: {len(os.listdir(directory_path))}")
    print(f"Total data extracted found: {data_found} and not found: {not_found}")
    print("###############################################################################################################################################################################################################################################################################################################################################")
print("+++++++++++++++++++++++++++++++++++++++++++++++++++++++++++")
print("Data Found", all_found)
print("Data Not Found", all_not_found)
print(pdf_v_dicts)
```
