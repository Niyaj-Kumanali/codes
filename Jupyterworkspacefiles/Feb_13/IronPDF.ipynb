```python
def extract_tables(pdf):
    table_rows_data = []

    data_to_exclude = re.compile(r'AMENDED PURCHASE ORDER Number (.+)\r\nRev : (.+)\r\n Page No: (.+) CESS\(INR\)',re.DOTALL)
    text_excluded = re.sub(data_to_exclude, "", text)
    table_data_matched = re.compile(r"CESS\(INR\) (.+)Grand Total", re.DOTALL).search(text_excluded)
    if table_data_matched:

        table_data_splitted = table_data_matched.group(1).strip().split('______________________________________________________________________________________________________________')
        table_data = [data for data in table_data_splitted if len(data)>3]
        for row in table_data:
            row_pattern = re.compile(r"\b(\d+)\s(\d+)\s(\d+)\s(.+?(?:PET|BTL|TUBE|WRP|CBD|CON))\s(\d+)\s(\w+)\s(\d+\.\d+)\s(\d+,\d+\.\d+)\s(\d+\.\d+)\s(\d+\.\d+)\s(\d+\,\d+\.\d+)\s\n(\d+)\s(\d+)?\s?(\d+\.\d+\.\d+)\s(\d+\.\d+)\s(\d+\.\d+)\s\n(\w+)\s\n\s(\d+)\s(\w+)\s(\d+)\b")

            matches = row_pattern.findall(row)

            formatted_matches = []
            for match in matches:
                elements = []
                for element in match:
                    elements.append(element)
                formatted_matches.append(elements)
            for match in formatted_matches:
                row_dict = dict()
                row_dict["Article No."] = int(match[1])
                row_dict["HSN/SAC Code"] = int(match[11])

                row_dict["EAN No"] = int(match[2])
                row_dict["Vendor Article No"] = int(match[12]) if len(match[12])>2 else None
                row_dict["Material Description"] = match[3]
                row_dict["Delivery Date"] = match[13] if len(match[13]) == 10 else match[12]+match[13]
                row_dict["Site"] = match[16]

                row_dict["Quantity"] = [match[4],match[17]]
                row_dict["UOM"] = [match[5],match[18]]
                row_dict["MRP"] = [match[6],match[19]]
                row_dict["Base Cost"] = float(''.join(match[7].split(",")))
                row_dict["CGST(%)"] = float(match[8])
                row_dict["SGST(%)"] = float(match[14])

                row_dict["CGST(INR)"] = float(match[9])
                row_dict["SGST(INR)"] = float(match[15])
                row_dict["Total Base  Value(INR)"] = float(''.join(match[10].split(",")))
                table_rows_data.append(row_dict)




    else:
        print("No Data Found")


    table_data_df = pd.DataFrame(table_rows_data)
    table_data_df["Delivery Date"] = pd.to_datetime(table_data_df["Delivery Date"])
    return table_data_df
```
