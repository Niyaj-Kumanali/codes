```python
from PyPDF2 import PdfReader
import pandas as pd
```

```python
reader = PdfReader(r"C:\Users\TP_USER\Desktop\AIML\Iron_pdf\cocacola.pdf")
page = reader.pages[0]
pdfdata=page.extract_text(0,90)
```

```python
pdfdata
```

```python
d=dict()
lines=pdfdata.split("\n")
lines
```

```python
flag=0
c=0
d["Item_Name"]=[]
d["Item_id"],d["Qty"],d["Price"],d["Amount"],d["descr"]=[],[],[],[],[]
for i in lines:
    if flag==1:
        try:
            if c==1:
                d["Item_Name"].append(i)
                c+=1
                continue
            elif c==2:
                temp=i.split()
                print(temp)
                d["Item_id"].append(temp[0])
                d["Qty"].append(temp[1])
                d["Price"].append(temp[2])
                d["Amount"].append(temp[3])
                c+=1
#                 print(d)
                continue
            elif c==3:
                d["descr"].append(i)
                c=1
        except:
            d["Item_Name"]=d["Item_Name"][:-1]
            d["Item_id"]=d["Item_id"][:-1]
            d["Qty"]=d["Qty"][:-1]
            break
            
    if "Amount" in i:
        flag=1
        c+=1
        
```
