```python
def format_dataframes(firstCry_df):
    for col in firstCry_df.columns.tolist():
        column = col
        for index in range(len(firstCry_df[col])):
            if not pd.isna(firstCry_df[col].iloc[index]):
                column += str(firstCry_df[col].iloc[index])
                firstCry_df[col].iloc[index] = np.nan
            else:
                break
        firstCry_df.rename(columns={col: column}, inplace=True)
    firstCry_df.dropna(how='all', inplace=True)
    for index, row in firstCry_df.iterrows():
        for nextIndex, nextRow in firstCry_df.iterrows():
            if index < nextIndex:
                if not pd.isna(nextRow['Sr.No.']):
                    break
                for col in firstCry_df.columns[1:]:
                    if not pd.isna(nextRow[col]):
                        firstCry_df.at[index, col] = str(firstCry_df[col][index]) +" "+ str(nextRow[col])
                        firstCry_df.at[nextIndex, col] = np.nan
    firstCry_df.dropna(how='all', inplace=True)
    return firstCry_df
```
