```python
import pyodbc
import pandas as pd

# Connection parameters
server = 'TPSYS-2132'  # e.g., 'localhost' or '127.0.0.1'
database = 'DemandForecasting'
username = 'sa'
password = 'Admin@123'

# Create a connection string
conn_str = (
    f'DRIVER={{ODBC Driver 17 for SQL Server}};'
    f'SERVER={server};'
    f'DATABASE={database};'
    f'UID={username};'
    f'PWD={password}'
)

# Connect to the database
conn = pyodbc.connect(conn_str)
cursor = conn.cursor()

print("Connection successfull")
```

```python
# Read CSV file into a DataFrame
csv_file_path = r"D:\React\DemandForecasting\backend\data\visualization_data.csv"
req_cols = ['CFACode', 'ProductCode', 'week_start_date', 'Qty', 'Category', 'CFAPINCODE', 'PREVQUANTITY']

# Load only the required columns
df = pd.read_csv(csv_file_path, usecols=req_cols)
# df['id'] = range(len(df))
df['predicted_qty'] = df['PREVQUANTITY']

df['week_start_date'] = pd.to_datetime(df['week_start_date'])
df.sort_values('week_start_date', ascending=True, inplace=True)
```

```python
# Get column names from the DataFrame
columns = df.columns.tolist()
columns
```

```python


# Construct SQL INSERT statement dynamically
placeholders = ', '.join(['?' for _ in columns])
column_names = ', '.join(columns)
sql_insert = f'INSERT INTO demand_forecasting ({column_names}) VALUES ({placeholders})'

# Prepare data for insertion
data = df.values.tolist()

# Execute the insert operation
cursor.executemany(sql_insert, data)

# Commit the transaction
conn.commit()

# Close the connection
cursor.close()
conn.close()

print("Data inserted successfully.")
```
