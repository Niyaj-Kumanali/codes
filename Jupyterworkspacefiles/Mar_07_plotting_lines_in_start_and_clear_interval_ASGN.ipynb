```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
events = pd.read_csv("data/events.csv")
events["event_clear_timestamp"].fillna(events["event_clear_timestamp"].mode()[0], inplace=True)
events.tail()
```

```python
events.info()
```

```python
plt.figure(figsize=(10, 5))
for index, row in events[50:90].iterrows():
    plt.plot([row["event_start_timestamp"], row["event_clear_timestamp"]], [row["event_key"], row["event_key"]], marker=".")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
for index, row in events[50:90].iterrows():
    sns.lineplot(x = [row["event_start_timestamp"], row["event_clear_timestamp"]], y = [row["event_key"], row["event_key"]], marker="o")
plt.xticks(rotation=90)
plt.show()
```
