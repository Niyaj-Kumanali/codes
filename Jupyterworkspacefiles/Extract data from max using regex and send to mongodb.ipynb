```python
import pandas as pd
from PyPDF2 import PdfReader
import re
import os

def extract_data(pdf_path):
    pdf_read = PdfReader(pdf_path)
    text = '\n'.join([page.extract_text() for page in pdf_read.pages])
    print(text)
    pattern = r"(\d+)\n(HIM.+?)(\d{13})\n(\d+)\s*\n(\w{0,3})\s*(\d+)\s*(\d+)\s*(\d+\.?\d*)(\w+)-(\d+)%-(\d+\.\d+)\n(\w+)-(\d+)%-(\d+\.\d{2})(\d+\.?\d*)"
    rows = re.findall(pattern, text, re.DOTALL)
    new_rows = []
    for match in rows:
        match = list(match)
        if match[8]=='CGST':
            temp1 = match[8]
            temp2 = match[9]
            temp3 = match[10]
            match[8] = match[11]
            match[9] = match[12]
            match[10] = match[13]
            match[11] = temp1
            match[12] = temp2
            match[13] = temp3
        new_rows.append(match)
    cols = ["SKU Code", "SKU Desc", "EAN", "HSN", "AR", "Qty", "MRP", "Unit Cost", "SGST", "SGST Tax-Rate%", "SGST TaxAmt","CGST", "CGST Tax-Rate%", "CGST TaxAmt", "Total Value"]
    df = pd.DataFrame(new_rows, columns=cols)
    del df['CGST']
    del df['SGST']
    
    numerics = ["SKU Code", "EAN", "HSN", "Qty", "MRP", "Unit Cost", "SGST Tax-Rate%", "SGST TaxAmt", "CGST Tax-Rate%", "CGST TaxAmt", "Total Value"]
    for col in numerics:
        df[col] = pd.to_numeric(df[col], errors='coerce')
    return df
    
directory = r"C:\Users\TP_USER\Downloads\POReaderFormats\POReaderFormats\Max"
all_dfs = []
for file in os.listdir(directory):
    pdf_path = os.path.join(directory, file)
    print(pdf_path)
    df = extract_data(pdf_path)
    all_dfs.append(df)
#     print(df.to_string())
#     print("____________________________________________________________")
merged_table = pd.concat([table for table in all_dfs], ignore_index=True)
merged_table
```

```python
'''
Delivery 
Address:MARINA MALL CHN
 MARINA MALL, NO 13/1A
EGATTUR OLD 
MAHABALIPURAM, 
TIRUPORUR TALUK  
KANCHIPURAM DISTRICT 
CHENNAI 603106PO No: 11237019 Supplier: 60009 - THE HIMALAYA DRUG COMPANY-
CHENNAI
PO Date: 07-APR-2024 100A ALAPAKKAM MAIN ROAD  
ALAPAKKAM PORUR   CHENNAI 600116 
GST: 33AAECM0847J1ZJ Delivery Date: 13-APR-2024 Ph.No: 9016157766
TIN: 33AAECM0847J1ZJ PO Expiry Date: 14-APR-2024 GST: 33AADFT3025B1ZW 
PO version: Original PO Status: Approved TIN:
'''

'''
Delivery 
Address:VEGAS MALL
VEGAS MALL  UNIT NOS LG
14, LOWER GROUND FLOOR,
PLOT NO. 06, SECTOR 14 
NORTH, DWARKA, NEW 
DELHI 110078PO No: 11236944 Supplier: 59770 - THE HIMALAYA DRUG COMPANY-
DELHI
PO Date: 07-APR-2024 915-916  RITHALA INDUSTRIAL AREA, 
RITHALA  NORTH WEST DELHI DELHI 
110085 
GST: 07AAECM0847J1ZE Delivery Date: 13-APR-2024 Ph.No: 9016157766
TIN: PO Expiry Date: 14-APR-2024 GST: 07AADFT3025B1ZR 
NA PO version: Original PO Status: Approved TIN:
'''

def extract_head_data(text):
    head_data = {}
    pattern = r"Delivery\s*\n?Address:(.+?)PO No:\s*(\d+)\s*Supplier:\s*(.+?)PO Date:\s*(\d+-\w+-\d+)(.+?)GST:\s*(\w*)\s*Delivery Date:\s*(\d+-\w+-\d+)\s*Ph.No:\s*(\d+)\n?\s*TIN:\s*(\w*)\s*PO Expiry Date:\s*(\d+-\w+-\d+)\s*GST:\s*(\w+)\s*\n(\w*)\s*PO version:\s*(\w+)\s*PO Status:\s*(\w+)\s*TIN:"
    matched = re.findall(pattern, text, re.DOTALL)
    if matched:
        for match in matched:
            head_data['Delivery Address'] = "".join(match[0].split('\n'))
            head_data['PO NO'] = match[1]
            head_data['Supplier'] = "".join("\n".join([match[2], match[4]]).split('\n'))
            head_data['PO Date'] = match[3]
            head_data['GST No'] = match[5]
            head_data['Delivery Date'] = match[6]
            head_data['Supplier Phone No'] = match[7]
            head_data['TIN'] = match[8]
            head_data['PO Expiry Date'] = match[9]
            head_data['Supplier GST No'] = match[10]
            head_data['PO Version'] = match[12]
            head_data['PO Status'] = match[13]
            head_data['Supplier TIN'] = match[11]
        df = pd.DataFrame(head_data, index=[0])
        dates = ['PO Date', 'Delivery Date', 'PO Expiry Date']
        for col in dates:
            df[col] = pd.to_datetime(df[col], errors='coerce')
        return df
    return None

        
# pdf_path = r"C:\Users\TP_USER\Downloads\POReaderFormats\POReaderFormats\Max\1712740152_2000072315020240410024041PM.pdf"
pdf_dir = r"C:\Users\TP_USER\Downloads\POReaderFormats\POReaderFormats\Max"
count = 1
for pdf in os.listdir(pdf_dir):
    pdf_path = os.path.join(pdf_dir, pdf)
    pdf_read = PdfReader(pdf_path)
    text = '\n'.join([page.extract_text() for page in pdf_read.pages])
    head_data_df = extract_head_data(text)
    if head_data_df is not None:
        count = upload_to_mssql(head_data_df, count)
        print("Successfully upload data into mssql")
    else:
        print("Error uploading data into mssql")
```

```python
import pandas as pd
import pyodbc
from datetime import datetime

def upload_to_mssql(df, count=1):
    # Check if DataFrame is empty
    if df.empty:
        print("DataFrame is empty. No records to upload.")
        return

    current_datetime = datetime.now()

    # Database connection parameters
    server = 'TPSYS-2132'
    database = 'Himalaya_PDF_data'
    username = 'sa'
    password = 'Admin@123'
    driver = '{ODBC Driver 17 for SQL Server}'

    # Create a connection string
    conn_str = f'DRIVER={driver};SERVER={server};DATABASE={database};UID={username};PWD={password}'

    try:
        # Connect to SQL Server
        conn = pyodbc.connect(conn_str)

        # Create a cursor
        cursor = conn.cursor()
        # Insert DataFrame records into SQL Server table
        for index, row in df.iterrows():
            cursor.execute("""
            INSERT INTO HPOF_MT_Order_Header (
                Customer, Upload_Order_No, Upload_Order_date, CustomerPONo,
                CustomerPODate, CustomerDeliveryDate, SOOverAllStatus,
                SOOverAllLastModified_on
            )
            VALUES (?, ?, ?, ?, ?, ?, ?, ?)
            """, 
            "Cust"+str(count), count, current_datetime,
            row['PO NO'], row['PO Date'], row['Delivery Date'],
            'S', current_datetime)
            count += 1

        # Commit the transaction
        conn.commit()
        print("Data uploaded successfully.")
        return count

    except Exception as e:
        print(f"Error uploading data: {e}")

    finally:
        # Close the cursor and connection
        cursor.close()
        conn.close()


```
