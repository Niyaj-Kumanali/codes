## High Pass Filters
* Sharpen an image
* Enhance high frequency parts of an image
* But contain noise so that first we pass the image in low pass filter

1. Convert the image into gray scale
2. Create custom Kernel
3. It is a Sobel Filter
4. kernel = np.array([[-1, 0, 1],[-2, 0, 2],[-1, 0, 1]])
5. Perform Convulution using filter2D, which has inputs: (grayscale image, bit-length ie -1, kernel)
6. Create threshold that sets all the filtered pixels to white above the threshold


```python
import cv2
import numpy as np

image = cv2.imread('image.jpg')

# 1. Convert the image into gray scale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# 2. Create the custom Kernel ie Sobel Filter
kernel = np.array([   [-1, 0, 1],
                      [-1, 0, 1],
                      [-1, 0, 1]    ])


# 3. Perform Convolution using filter2D, which has inputs: (grayscale image, bit-length ie -1, kernel)

filtered_image = cv2.filter2D(gray_image, -1, kernel)


# 4. Create threshold that sets all the filtered pixels to white above the threshold

threshold_image = cv2.threshold(filtered_image, )


```
