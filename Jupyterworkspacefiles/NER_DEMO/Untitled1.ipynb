```python
import fitz  # PyMuPDF

def extract_text_from_pdf(pdf_file):
    # Open the PDF file
    pdf_document = fitz.open(pdf_file)
    
    text = ""
    for page_num in range(pdf_document.page_count):
        # Extract text from each page
        page = pdf_document.load_page(page_num)
        text += page.get_text()
    
    return text

def detect_tables(text):
    # Implement logic to detect tables
    # You can use regular expressions or other heuristics
    # For simplicity, let's assume we're detecting tables based on lines
    lines = text.split('\n')
    tables = []
    table = ""
    for line in lines:
        if line.strip().startswith("|") or line.strip().endswith("|"):
            table += line + "\n"
        elif table:
            tables.append(table.strip())
            table = ""
    if table:
        tables.append(table.strip())
    return tables

def generate_latex_from_tables(tables):
    # Convert detected tables to LaTeX tabular environment
    latex_content = ""
    for table in tables:
        latex_content += "\\begin{table}\n"
        latex_content += "\\begin{tabular}{|c|c|}\n"
        latex_content += "\\hline\n"
        rows = table.split("\n")
        for row in rows:
            cols = row.split("|")
            cols = [col.strip() for col in cols if col.strip()]
            latex_content += " & ".join(cols) + " \\\\\n"
        latex_content += "\\hline\n"
        latex_content += "\\end{tabular}\n"
        latex_content += "\\end{table}\n"
    return latex_content

def generate_latex_from_text(text):
    # Enclose the text in a LaTeX document structure
    latex_content = "\\documentclass{article}\n"
    latex_content += "\\begin{document}\n"
    latex_content += text
    latex_content += "\\end{document}"
    
    return latex_content

def save_latex_to_file(latex_content, output_file):
    with open(output_file, "w", encoding='utf-8') as f:
        f.write(latex_content)

def pdf_to_latex(pdf_file, output_file):
    # Extract text from the PDF
    text = extract_text_from_pdf(pdf_file)
    
    # Detect tables in the extracted text
    tables = detect_tables(text)
    
    # Generate LaTeX content from the detected tables
    latex_tables = generate_latex_from_tables(tables)
    
    # Generate LaTeX content from the remaining text
    remaining_text = text
    for table in tables:
        remaining_text = remaining_text.replace(table, "")
    latex_text = generate_latex_from_text(remaining_text)
    
    # Combine LaTeX content
    latex_content = latex_text + "\n" + latex_tables
    
    # Save the LaTeX content to a file
    save_latex_to_file(latex_content, output_file)
    
    print("PDF converted to LaTeX successfully.")

# Example usage
pdf_file = r"D:/Downloads/PO_format_pdfs/Walmart/1712733039_24000070938720240410124200PM.PDF"
output_file = "output.tex"
pdf_to_latex(pdf_file, output_file)

```
