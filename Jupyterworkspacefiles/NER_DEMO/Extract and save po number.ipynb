```python
import re
from PyPDF2 import PdfReader
import os

def extract_po_number(pdf_path):
    text = ''
    try:
        pdf_read = PdfReader(pdf_path)
        text = '\n'.join([page.extract_text() for page in pdf_read.pages])
    except Exception as e:
        print(f"Error processing {pdf_path}: {e}")
        return None

    walmart = r"(\d{10}\s*)PURCHASE ORDER"
    vishal = r"PO Number:\s*(\d{10})"
    reliance = r"PO NO\.:\s*(\d{10})"
    reliance2 = r"Purchase Order:\n?\s*(\d{10})"
    optival = r"Purchase Order No:\s*(\d*)"
    more = r"PO Number:\s*(\d*)"
    metro1 = r"PO NO\.:\s*(\d{10})"
    metro2 = r"Order no.:\s*(\d{9})"
    metro3 = r"Art\. no\d{2}\.\d{2}\.\d{4}(\d*)"
    max = r"PO No:\s*(\d*)"
    first_cry = r"PONO:\s*(\w*/?\w*/?\d*)"
    avenue_supermarts = r"PO #.+?(\d+)"
    apollo = r"PO No\.\s*:.+?(PO-\w+)"
    aditya = r"PO Number:\s*(\d*)"
    pattern = f"{walmart}|{vishal}|{reliance}|{reliance2}|{optival}|{more}|{metro1}|{metro2}|{metro3}|{max}|{first_cry}|{avenue_supermarts}|{apollo}|{aditya}"
    matches = re.findall(pattern, text, re.DOTALL)
    po_numbers = set()

    for match in matches:
        for found in match:
            if found.strip() != '':
                po_numbers.add(found.strip())

    if po_numbers:
        return list(po_numbers)
    return None
```

```python
pdf_dir = r"C:\Users\TP_USER\Downloads\POReaderFormats\POReaderFormats"
po_nums = set()
for folder in os.listdir(pdf_dir):
    pdf_folder = os.path.join(pdf_dir, folder)
    for pdf in os.listdir(pdf_folder):
        if not pdf.lower().endswith("pdf"):
            continue
        pdf_path = os.path.join(pdf_folder, pdf)
        ponum = extract_po_number(pdf_path)
        if ponum is None:
            continue
        for po in ponum:
            po_nums.add(po)
po_nums
```
