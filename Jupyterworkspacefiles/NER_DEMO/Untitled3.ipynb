```python
import re
def clean_text(text):
    for i in range(3):
        text = re.sub(r"\n+", " ", text)
        text = re.sub(r"\s+", " ", text)
        text = re.sub(r"[^a-zA-Z0-9\s.,?!@#$%^&*()_+=\-`~{}[\]:;'\"|/\\<>]", " ", text)
        text = re.sub(r"(\d{2}|\d{4})[./](\d{2})[./](\d{4}|\d{2})", " \g<1>-\g<2>-\g<3> ", text) #to change date format to - format
        text = re.sub(r"(\d+\.\d{2})", "\g<1> ", text) # to sepeate the prices
        text = re.sub(r"([A-Z]{5}\d{4}[A-Z])", " \g<1> ", text) #seperate the pan
        text = re.sub(r"(\d{2})\s*([A-Z]{5}[0-9]{4}[A-Z])\s*([0-9][0-9A-Z][0-9A-Z])", " \g<1>\g<2>\g<3> ", text) #add and seperate the gstin
        text = re.sub(r"\((.+?)\)", r" ( \g<1> ) ", text) #separete values within brackets
        text = re.sub(r"\[(.+?)\]", r" [ \g<1> ] ", text) #separete values within brackets
        text = re.sub(r"(Him.+?\d+\s*[a-zA-Z.]*\s*[a-zA-Z.]*)", " \g<1> ", text, flags=re.I) #separate product description
        text = re.sub(r"\s*%\s*", "% ", text) # remove extra space between % and add it after
        text = re.sub(r"(\d+%)", " \g<1> ", text) # add spaces on both side of percent values
        text = re.sub(r":", " : ", text) # add spaces on both sides
        text = re.sub(r"([^a-zA-Z0-9])\1+", r"\1", text) # remove extra symbols
        text = re.sub(r"([#/])", " \g<1> ", text) # add spaces around this characters
        text = re.sub(r"(\d{2})\s*:\s*(\d{2})\s*:\s*(\d{2})\s*", " \g<1>:\g<2>:\g<3> ", text) # correct time format removing extra spaces
        text = re.sub(r"(Dt|Vendor|PO )", " \g<1> ", text, flags=re.I)
        text = re.sub("\s+", " ", text) # remove extra spaces
    return text.strip()

clean_text("")
```

```python
import json
import spacy
import json
from spacy.training import offsets_to_biluo_tags
import os

def check_data(nlp, train_data):
    flag = False
    for text, entities in train_data:
        # print(text, entities)
        try:
            tags = offsets_to_biluo_tags(nlp.make_doc(text), entities['entities'])
            if "-" in tags:
                print(f"Tags for '{text}' --> {tags}")
                flag = True
        except:
            print(text[:500], entities)
            
    if flag:
        print("Mislabelled Training Data")
    else:
        print("Correctly labelled training data")
        
pdfs_dir = r"C:\Users\TP_USER\Downloads\POReaderFormats\POReaderFormats"
train_data = []
for folder in os.listdir(pdfs_dir):
    pdf_dir = os.path.join(pdfs_dir, folder)
    for pdf_name in os.listdir(pdf_dir):
            if pdf_name.lower().endswith("json") and folder == 'AvenueSupermarts':
            # Write text to the file
                path = os.path.join(pdf_dir, pdf_name)
                with open(path, "r", encoding="utf-8") as f:
                    data = json.load(f)
                if data['annotations']:
                    train_data.append(data['annotations'][0])
nlp = spacy.load("en_core_web_sm")
check_data(nlp, train_data)
```
