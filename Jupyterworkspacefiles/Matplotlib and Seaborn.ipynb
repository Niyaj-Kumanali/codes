```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = np.random.randint(1,10, 100)
y = np.random.randint(1,10, 100)
```

```python
df = pd.DataFrame({"x":x, "y":y, "result": x<y})
df
```

```python
plt.scatter(df["x"],df["y"])
plt.grid(True)
```

```python
sns.set_style("whitegrid")
sns.FacetGrid(df, hue="result").map(plt.scatter, "x", "y").add_legend()
```

```python
titanic = sns.load_dataset('titanic')
titanic
```

```python
sns.countplot(x="who",hue='sex', data=titanic)
plt.grid(True)
```

```python
plt.scatter(titanic[titanic['alone'] == False]['fare'], titanic[titanic['alone'] == False]['age'])
plt.scatter(titanic[titanic['alone'] == True]['fare'], titanic[titanic['alone'] == True]['age'])
```

```python
sns.scatterplot(x="fare", y="age", hue="alone",data=titanic)
```
